```python
import network
import activation
import loss
```

```python
import numpy as np
```

Testing the network can learn to create a summation function

```python
nn = network.Network([5, 1], [activation.Relu], loss.MSE, 0.3)
```

```python
data = np.random.rand(1000, 5)
labels = np.sum(data, axis = 1)[np.newaxis].T
```

```python
labels
```

```python
data[:5]
```

```python
losses = nn.train(data, labels)
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.ylim([0, 1])
plt.show()
```

```python
test_data = np.random.rand(400, 5)
test_labels = np.sum(test_data, axis = 1)[np.newaxis].T
```

```python
performance = nn.test(test_data, test_labels)
plt.plot(performance)
plt.show()
```

Testing the network can learn to create a boolean function

```python
data = np.random.choice([0, 1], size = (50000, 2))
data[:5]
```

```python
labels = data.any(axis = 1)[np.newaxis].T.astype(float)
labels[:5]
```

```python
ornn = network.Network([2, 4, 1], [activation.Relu, activation.Relu], loss.MSE, 0.03)
```

```python
l = ornn.train(data, labels)
```

```python
plt.plot(l)
plt.show()
```

```python
subject = np.random.choice([0, 1], size = (1, 2))
subject
```

```python
ornn.predict(subject)
```
